```python
## Tools zum durchsuchen eines exports aus label-studio im JSON Format
import json
import time

path_to_json = "project-1-at-2023-01-23-15-16-4b8bba1d.json"


def extract_test_json_for_labels(label):
    result = []
    result = result + extract_test_json_for_label(path_to_json, label, result)
    result = list(set(result))

    return result


def extract_test_json_for_label(path_to_json, label, result):
    json_file = open(path_to_json)
    json_data = json.loads(json_file.read())

    for item in json_data:
        for annotation in item["annotations"]:
            for rect in annotation["result"]:
                if label in rect["value"]["rectanglelabels"] and item["file_upload"] not in result:
                    image_name = item["file_upload"]
                    result.append(image_name[9:])

    return result


label_for_csv = "child"
name_of_all_labels_images = extract_test_json_for_labels(label_for_csv)
print(name_of_all_labels_images)

import pandas as pd

df = pd.DataFrame(name_of_all_labels_images)
df.to_csv('childlist ' + str(time.asctime()) + '.csv', header=None)
```
